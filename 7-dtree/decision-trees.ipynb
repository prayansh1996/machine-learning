```python
ytrain = data['Survived'].values #Survived or not
xtrain = npa([data['Pclass'], data['Sex'], data['Age'], # Pclass	Sex	Age	SibSp	Parch	Fare
              data['SibSp'], data['Parch'], data['Fare'], 
              data['Survived']]).astype("float").T  
label = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']
index = { 'Pclass':0, 'Sex':1, 'Age':2, 'SibSp':3, 'Parch':4, 'Fare':5, 'Survived':6}
print xtrain.shape#[:,-1]
```

```python
def entropy(data): #give it divided data, and it shall return its entropy
    p = 0.0
    for im in range(data.shape[0]): #could have straight away taken p = data[-1].mean() as values are only 0 and 1
        if data[im,-1] == 1:
            p += 1
    p /= data.shape[0]
    if p==0 or p==1:
        return 0
    entropy = -p*np.log2(p) - (1-p)*np.log2(1-p)
    return entropy
```

```python
def split(data, col, mean, disp = False):
    if disp == True:
        print "Mean->", mean, " for ", label[col]
    l,r = [], []
    for im in range(data.shape[0]):
        if data[im,col] <= mean:
            l.append(data[im])
        else:
            r.append(data[im])
    return npa(l), npa(r)

l, r = split(xtrain,index['Age'], xtrain[:,index['Age']].mean())
print l.shape, r.shape
```

```python
def informationGain(parent,l,r):
    return (entropy(parent) - (float(l.shape[0])/parent.shape[0])*entropy(l) - (float(r.shape[0])/parent.shape[0])*entropy(r))

print "parent:", entropy(xtrain)
print "l:\t", entropy(l)
print "r:\t", entropy(r)
print "gain:  ", informationGain(xtrain,l,r)
```

```python
class dtnode:
    def __init__(self, depth=0, maxDepth = 10000):
        self.col = 0
        self.l = None
        self.r = None
        self.splitVal = float('Nan')
        self.gain = 0.0
        self.survived = 0
        self.dead = 0
        self.depth = depth
        self.maxDepth = maxDepth
    
    def fit(self,data):        
        #fill dead and alive
        for im in range(data.shape[0]):
            if data[im,-1] == 1:
                self.survived += 1
            else:
                self.dead += 1
        
        #find col to be split on
        maxGain = -100 #-inf
        for col in range(data.shape[1]-1):
            mean = data[:,col].mean()
            l, r = split(data,col,mean)
            if len(l)==0 or len(r)==0:
                continue
                
            gain = informationGain(data,l,r)
            if gain > maxGain:
                maxGain = gain
                self.col = col
        
        if maxGain == -100:
            return
        
        self.splitVal = data[:,self.col].mean()
        l, r = split(data, self.col, self.splitVal)
        if len(l) == 0 or len(r) == 0 or self.depth >= self.maxDepth-1:
            return 
        
        else:
            self.l = dtnode(depth = self.depth+1, maxDepth = self.maxDepth)
            self.l.fit(l)
            
            self.r = dtnode(depth = self.depth+1, maxDepth = self.maxDepth)
            self.r.fit(r)   
    
    def decide(self):
        ans = 0
        if self.survived >= self.dead:
            ans = 1
        
        larger = max(self.survived, self.dead)
        smaller = min(self.survived, self.dead)
        return (ans, (larger,smaller))
    
    def predict(self, sample):
        if self.l == None and self.r == None:
            return self.decide()
        
        if sample[self.col] <= self.splitVal:
            return self.l.predict(sample)
        else:
            return self.r.predict(sample)
    
    def recurse(self, maxDepth = None):
        if (maxDepth != None) and (self.depth>=maxDepth):
            return
        
        for i in range(2*self.depth):
            print ".",
        print "\_", 
        print self.survived, "/", self.dead," split on -> ", label[self.col]
        
        if self.l is not None:
            self.l.recurse(maxDepth = maxDepth)
        if self.r is not None:
            self.r.recurse(maxDepth = maxDepth)
    
    
```
